```python
t = [2,4,6,8,10,12,14]
t_predict = [1.8]
alphas = [0.2, 0.4, 0.6, 0.8]
```

```python
def predict(alpha, t, n, t_predict):
  prediction = alpha*t[n-1] + (1-alpha) * t_predict[n-1]
  t_predict.append(prediction)
  return prediction
```

```python
import matplotlib.pyplot as plt

for alpha in alphas:
  for n in range(len(t)):
    if n == 0:
      continue
    prediction = predict(alpha=alpha, t=t, n=n, t_predict=t_predict)
  # print(t_predict)
  plt.plot(t, t_predict, marker='o', linestyle='-')
  plt.xlabel('True time (t)')
  plt.ylabel('Predicted time (toue)')
  plt.title(f'For alpha = {alpha}')
  plt.grid(True)
  plt.show()

  t_predict_0 = t_predict[0]
  t_predict.clear()
  t_predict.append(t_predict_0)
```
